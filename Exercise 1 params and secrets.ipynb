In this exercise, we learn how to use workflow parameters and secrets. These are used to set custom values to the workflow, without having to recontainerize the cells of the notebook.

We start with the tutorial workflow ([Your first workflow in Python.ipynb](./Your%20first%20workflow%20in%20Python.ipynb)), and modify it to add parameters and secrets. The cells are copied in this notebook, so that you can modify them directly.

_Tip:_ Read the [NaaVRE documentation about special variables](https://naavre.net/docs/NaaVRE_Interface/#special-variables).

_Step 1_

Create a code cell at the start of the notebook with two variables:

```python
param_greeting = "Hello"
secret_password = "don't tell anyone"
```

````python
# (DO NOT containerize this cell)

param_greeting = "Hello"
param_names = ["Alice", "Bob"]
secret_password = "don't tell anyone"
````

````python
# ex1 1: input list
names = param_names
````

_Step 2_

Modify cell 2 (`process list`) to use the new variables: 

```python
print(f"{param_greeting}, {name}! The password is: {secret_password}")
```

````python
# ex1 2: process list
for name in names:
    print(f"{param_greeting}, {name}! The password is: {secret_password}")
````

_Step 3_

Containerize the cells

_Step 4_

Create a workflow.

Run the workflow, and check that you get asked for values for `param_greeting` and `secret_password`
